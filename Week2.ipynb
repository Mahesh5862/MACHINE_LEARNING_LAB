```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv("Car_sales.csv")

print(df.isnull().sum())
#replace the dot with Nan
df["Horsepower"]=df["Horsepower"].replace('.', np.nan)
df["Price in thousands"]=df["Price in thousands"].replace('.', np.nan)







df
```

```python

# df["Price in thousands"] = pd.to_numeric(df["Price in thousands"], errors='coerce')

# Fill NaN values with the mean of the column
df["Price in thousands"] = df["Price in thousands"].fillna(df["Price in thousands"].mean())
df
```

```python
#sort the values according to price
df.sort_values(by="Price in thousands",ascending=True)
```
